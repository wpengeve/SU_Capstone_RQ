```python
# Columns to replace NaN values
cols_to_replace = ['20161','20162','20163','20164','20171','20172','20173','20174','20181','20182','20183','20184','20191','20192','20193','20194','20201','20202','20203','20204','20211','20212','20213','20214','20221','20222','20223','20224', 'Grand Total']

# Convert column names to strings
sales_data.columns = sales_data.columns.astype(str)

# Replace NaN values with 0 in the specified columns
sales_data[cols_to_replace] = sales_data[cols_to_replace].fillna(0)

# Replace negative values with 0 in the specified columns
sales_data[cols_to_replace] = sales_data[cols_to_replace].applymap(lambda x: max(0, x))


```

```python
from itertools import groupby
from operator import itemgetter

# Find the indices of the rows where 'ZXG ID' is blank
blank_indices = sales_data[sales_data['ZXG ID'].isna()].index.tolist()

# Function to convert list of indices to ranges
def indices_to_ranges(indices):
    """Convert a list of indices to ranges"""
    ranges = []
    for k, g in groupby(enumerate(indices), lambda i_x:i_x[0]-i_x[1]):
        consecutive_indices = list(map(itemgetter(1), g))
        if len(consecutive_indices) == 1:
            ranges.append(str(consecutive_indices[0]))
        else:
            ranges.append(f"{consecutive_indices[0]}-{consecutive_indices[-1]}")
    return ranges

# Convert indices to ranges
blank_ranges = indices_to_ranges(blank_indices)
blank_ranges
```

```python
# Find all the rows where 'Country' is blank
blank_rows_country = sales_data[sales_data['Country'].isna()].index.tolist()

# Apply function to get ranges
blank_ranges_country = indices_to_ranges(blank_rows_country)
blank_ranges_country
```

```python
# Replace the blanks in the specified ranges with the specified values
sales_data.loc[6934:7079, 'Country'] = sales_data.loc[6934:7079, 'Country'].fillna('Moe_Country')
sales_data.loc[62275:62294, 'Country'] = sales_data.loc[62275:62294, 'Country'].fillna('Wanchi_Country')

# Check if the replacements were successful
sales_data.loc[6934:7079]['Country'].unique(), sales_data.loc[62275:62294]['Country'].unique()
```

```python
# Initialize a company counter
company_counter = 1

# Iterate over the blank ranges in the 'ZXG ID' column
for start, end in [(int(r.split('-')[0]), int(r.split('-')[1])) if '-' in r else (int(r), int(r)) for r in blank_ranges]:
    # Get the unique countries in the current range
    unique_countries = sales_data.loc[start:end, 'Country'].unique()

    # Iterate over the unique countries
    for country in unique_countries:
        # Get the indices for the current country within the current range
        #indices = sales_data.loc[start:end][sales_data['Country'] == country].index
        indices = sales_data.loc[start:end][sales_data.loc[start:end, 'Country'] == country].index

        # Assign the company name to the 'ZXG ID' column for the current indices
        sales_data.loc[indices, 'ZXG ID'] = f'MW{company_counter}'

        # Increment the company counter
        company_counter += 1

# Check the first few rows of the modified dataset
sales_data.loc[[5780,5983,6934,7079,14962,15053,59837,59838,59839,59840,59841], :]
```

```python
# Convert 'Industry' to string type
sales_data['Industry'] = sales_data['Industry'].astype(str)

# Replace blank values with '99'
sales_data['Industry'].replace(['nan', ''], '99', inplace=True)

```

```python
# Remove rows with blank 'PA part description'
sales_data = sales_data.dropna(subset=['PA part description'])
```

```python
# Remove rows with the specified value in 'UT L30 - Segment' column
value_to_remove = 'Data Exchange Service Offering (Expert Labs Only)'
sales_data = sales_data[sales_data['UT L30 - Segment'] != value_to_remove]
```

```python
#Replace negative values with 0
sales_data[cols_to_replace] = sales_data[cols_to_replace].applymap(lambda x: 0 if x < 0 else x)
```

```python
# Preview the modified dataset to confirm the changes
sales_data.head()
```

```python
# Standardize the revenue data by replacing any value > 0 with 1, indicating a sale
# Columns representing the quarterly sales data are specified in cols_to_replace
sales_data[cols_to_replace] = sales_data[cols_to_replace].applymap(lambda x: 1 if x > 0 else 0)

# Preview the modified dataset to confirm the changes
sales_data.head()
```

```python
# Exclude the following column for long format transformation
id_vars = ['ZXG ID', 'Country', 'Industry', 'UT L30 - Segment', 'Family', 'Product name', 'PA part description', 'Rev type']
value_vars = sales_data.columns.drop(id_vars + ['Grand Total']).tolist()

# Melt the DataFrame to long format
sales_data_long = pd.melt(sales_data, id_vars=id_vars, value_vars=value_vars, var_name='Quarter', value_name='Revenue')

# Keep only the specified quarters
#quarters_to_keep = ['20221','20222','20223','20224','20171','20172','20173','20174']
#sales_data_long = sales_data_long[sales_data_long['Quarter'].isin(quarters_to_keep)]

# Filter for 'Industry' = 1
#sales_data_long = sales_data_long[sales_data_long['Industry'] == '1.0']

# Keep only the rows where a sale occurred, i.e., 'Revenue' = 1
#sales_data_long = sales_data_long[sales_data_long['Revenue'] == 1]

# Keep only the rows where a Rev type = 'Subs and Supp'
#sales_data_long = sales_data_long[sales_data_long['Rev type'] == 'Subs and Supp']

# Keep only the rows where a Product name = 'MFT Connect:Direct'
#sales_data_long = sales_data_long[sales_data_long['Product name'] == 'MFT Connect:Direct']

# Keep only the rows for United States
#sales_data_long = sales_data_long[sales_data_long['Country'] == 'United States']

# Drop unnecessary columns, but retain 'Product name' for grouping
cols_to_drop = ['Country', 'Industry', 'UT L30 - Segment', 'Family', 'Rev type']
sales_data_long.drop(columns=cols_to_drop, inplace=True)

# Convert the quarter to datetime format
def quarter_to_datetime(quarter_str):
    year = quarter_str[:4]
    q = int(quarter_str[4])
    month = (q - 1) * 3 + 1  # Get the starting month of the quarter
    return f"{year}{month:02}"

sales_data_long['Quarter'] = sales_data_long['Quarter'].apply(quarter_to_datetime)
sales_data_long['Quarter'] = pd.to_datetime(sales_data_long['Quarter'], format='%Y%m')

# Sort the DataFrame by company, product name, and quarter
sales_data_long.sort_values(['ZXG ID', 'Product name', 'Quarter'], inplace=True)

# Reset the index
sales_data_long.reset_index(drop=True, inplace=True)

# Function to remove consecutive duplicates
def remove_consecutive_duplicates(seq):
    if not seq:
        return []

    result = [seq[0]]
    for item in seq[1:]:
        if item != result[-1]:
            result.append(item)
    return result

# Group by 'ZXG ID' and 'Product name' and then remove consecutive duplicates
grouped_sequences = sales_data_long.groupby(['ZXG ID', 'Product name','Quarter'])['PA part description'].apply(list)
grouped_sequences = grouped_sequences.apply(remove_consecutive_duplicates)

# Convert to a list of lists for the PrefixSpan algorithm
grouped_sequences_list = [list(group) for _, group in grouped_sequences.items()]

```

```python
def top_three_before_and_after(product_name, sequences):
    preceding_products_count = {}
    subsequent_products_count = {}

    # Process sequences
    for seq in sequences:
        if product_name in seq:
            index = seq.index(product_name)

            # Check for preceding product
            if index > 0:
                preceding_product = seq[index - 1]
                preceding_products_count[preceding_product] = preceding_products_count.get(preceding_product, 0) + 1

            # Check for subsequent product
            if index + 1 < len(seq):
                subsequent_product = seq[index + 1]
                subsequent_products_count[subsequent_product] = subsequent_products_count.get(subsequent_product, 0) + 1

    # Sort and retrieve top three preceding and subsequent products
    top_three_preceding = sorted(preceding_products_count.items(), key=lambda x: x[1], reverse=True)[:3]
    top_three_subsequent = sorted(subsequent_products_count.items(), key=lambda x: x[1], reverse=True)[:3]

    # Display results
    print(f"For the product '{product_name}':")
    for product, frequency in top_three_preceding:
        print(f"Before buying, the product '{product}' was bought {frequency} times.")
    for product, frequency in top_three_subsequent:
        print(f"After buying, the product '{product}' was bought {frequency} times.")

# Usage
product_name_input = 'STERLING C D PREM ED PVU SUB'  # Change this to any product name of interest
top_three_before_and_after(product_name_input, grouped_sequences)
```

```python
from collections import Counter

def most_frequent_baskets(product_name, sequences, top_n=2):
    baskets_with_product = []

    # Find all sequences containing the product
    for seq in sequences:
        if product_name in seq:
            baskets_with_product.append(tuple(sorted(seq)))  # Sort to ensure each basket has a unique representation

    # Count the frequency of each unique basket
    basket_counter = Counter(baskets_with_product)

    # Sort baskets by frequency
    sorted_baskets = sorted(basket_counter.items(), key=lambda x: x[1], reverse=True)

    # Display the top_n most frequent baskets
    print(f"Top {top_n} most frequent baskets containing '{product_name}':")
    for basket, frequency in sorted_baskets[:top_n]:
        print(f"Basket {basket} appears {frequency} times.")

product_name_input = 'STERLING C D PREM ED PVU SUB'
most_frequent_baskets(product_name_input, grouped_sequences)
```
